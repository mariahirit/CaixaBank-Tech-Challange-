Aislamos tus datos "outliers"

```python
data = data.query('Open>2000')
data
```

```python
data ['Date'] = pd.to_numeric(data['Date'], errors='coerce')
data = data.replace(np.nan, 0, regex=True)
```

```python
data
```

FASE 2. Entrenamiento del modelo


```python
X=data.Close[::, np.newaxis]
y=data.Target
```

```python
from sklearn.model_selection import train_test_split
```

Algoritmos
#LogisticRegression


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
from sklearn.preprocessing import StandardScaler
escalar = StandardScaler()

X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)

from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
algoritmo.fit(X_train, y_train)
```

```python
y_pred = algoritmo.predict(X_test)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.show()
```

Para averiguar la precisión, la sensibilidad  del algoritmo utilizamos la librería metrics de scikit-learn e importamos precision_score, y la implementamos junto con los datos reales y los que hemos obtenidos utilizando el modelo.

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred)
print('Precisión del modelo:')
print(precision)
```

```python
from sklearn.metrics import accuracy_score
exactitud = accuracy_score(y_test, y_pred)
print('Exactitud del modelo:')
print(exactitud)
```

```python
from sklearn.metrics import recall_score
sensibilidad = recall_score(y_test, y_pred)
print('Sensibilidad del modelo:')
print(sensibilidad)
```

```python
from sklearn.metrics import f1_score
puntajef1 = f1_score(y_test, y_pred)
print('Puntaje F1 del modelo:')
print(puntajef1)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_test, y_pred)
print('Curva ROC - AUC del modelo:')
print(roc_auc)
print('Precisión del modelo:', precision)
print('Exactitud del modelo:', exactitud)
print('Sensibilidad del modelo:', sensibilidad)
print('Puntaje F1 del modelo:', puntajef1)
print('Curva ROC - AUC del modelo:', roc_auc)
```
